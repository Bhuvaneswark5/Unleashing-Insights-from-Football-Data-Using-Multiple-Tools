# 1.What is the probability of a player scoring a goal given that they received a yellow card?

```python
# Calculate total number of players with yellow cards
yellow_card_players = df_merged[df_merged['yellow_cards'] > 0]

# Calculate total goals scored by players with yellow cards
goals_with_yellow_cards = yellow_card_players[yellow_card_players['goals'] > 0].shape[0]

# Total players with yellow cards
total_yellow_card_players = yellow_card_players.shape[0]

# Conditional Probability: P(Goal | Yellow Card)
if total_yellow_card_players > 0:
    p_goal_given_yellow = goals_with_yellow_cards / total_yellow_card_players
else:
    p_goal_given_yellow = 0

print(f'Probability of scoring a goal given a yellow card: {p_goal_given_yellow:.2f}')
```

# 2.What is the joint probability of a player being a Defender and receiving a yellow card

```python
# Calculate total number of Defenders
defenders = df_merged[df_merged['position_x'] == 'Defender']

# Calculate total yellow cards received by Defenders
yellow_cards_for_defenders = defenders[defenders['yellow_cards'] > 0].shape[0]

# Total number of players
total_players = df_merged.shape[0]

# Joint Probability: P(Defender and Yellow Card)
if total_players > 0:
    p_defender_and_yellow = yellow_cards_for_defenders / total_players
else:
    p_defender_and_yellow = 0

print(f'Joint Probability of being a Defender and receiving a yellow card: {p_defender_and_yellow:.2f}')
```

# How does the distribution of goals scored by players in our sample compare to the overall population, and what implications does this have for analyzing player performance?"

```python
import numpy as np

# Population mean of goals scored
population_mean_goals = df_merged['goals'].mean()

# Randomly sample 30 players
sample_size = 30
sample = df_merged.sample(n=sample_size, random_state=42)
sample_mean_goals = sample['goals'].mean()

print(f'Population Mean Goals: {population_mean_goals:.2f}')
print(f'Sample Mean Goals (n={sample_size}): {sample_mean_goals:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Plotting the distribution of goals scored
plt.figure(figsize=(10, 6))
sns.histplot(df_merged['goals'], bins=10, kde=True)
plt.title('Distribution of Goals Scored by Players')
plt.xlabel('Goals Scored')
plt.ylabel('Frequency')
plt.show()

# Perform Shapiro-Wilk test for normality
shapiro_test = stats.shapiro(df_merged['goals'])
print(f'Shapiro-Wilk Test Statistic: {shapiro_test.statistic:.4f}, p-value: {shapiro_test.pvalue:.4f}')
```

```python
# Calculate sample mean and standard deviation
sample_mean = df_merged['goals'].mean()
sample_std = df_merged['goals'].std()

# Calculate the standard error
standard_error = sample_std / np.sqrt(len(df_merged))

# Calculate the margin of error for a 95% confidence interval
z_score = stats.norm.ppf(0.975)  # 95% confidence
margin_of_error = z_score * standard_error

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f'95% Confidence Interval for Average Goals Scored: {confidence_interval[0]:.2f} to {confidence_interval[1]:.2f}')
```

interpretation:Overall, while the average goals scored in our sample is lower than the population mean, 
the data does not appear to follow a normal distribution, which may affect our analysis. 
The confidence interval gives us a range where we expect the true population mean to fall.

# Question 1: Can we predict the outcome of a match (win, loss, or draw) for a team based on player performance and match conditions?

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load your DataFrame (assuming itâ€™s named df_merged)
df = df_merged.copy()  # Adjust this line as needed
```

```python
# Create the target variable for match outcomes
df['match_outcome'] = np.where(df['home_club_goals'] > df['away_club_goals'], 'win',
                                np.where(df['home_club_goals'] < df['away_club_goals'], 'loss', 'draw'))

# Keep only relevant columns for analysis
features = df[['home_club_goals', 'away_club_goals', 'yellow_cards', 'red_cards', 'attendance', 'competition_id_x']]
target = df['match_outcome']
```

```python
# One-hot encode categorical features
features = pd.get_dummies(features, columns=['competition_id_x'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Classification Report
print(classification_report(y_test, y_pred))

# Confusion Matrix
print(confusion_matrix(y_test, y_pred))

# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

##The model achieves near-perfect performance on the testing data. It correctly classifies all instances into their respective classes (draw, loss, win). This indicates that the model is highly accurate and reliable in predicting match outcomes based on the given features.

# hypothesis testing using t test

Is there a significant difference in the average number of goals scored by the top five teams compared to the bottom five teams in the league?

Step 1: Define Hypotheses
Null Hypothesis (H0): There is no significant difference in the average number of goals scored by the top five teams and the bottom five teams.
u top=ubottom
 

Alternative Hypothesis (H1): There is a significant difference in the average number of goals scored by the top five teams and the bottom five teams.
u top not equal to u bottom

```python
# Calculate average goals scored by each team
team_performance = df_merged.groupby('home_club_name')['goals'].mean().reset_index()

# Identify the top 5 and bottom 5 teams based on average goals
top_teams = team_performance.nlargest(5, 'goals')['home_club_name']
bottom_teams = team_performance.nsmallest(5, 'goals')['home_club_name']

# Filter the original DataFrame for these teams
top_goals = df_merged[df_merged['home_club_name'].isin(top_teams)]['goals']
bottom_goals = df_merged[df_merged['home_club_name'].isin(bottom_teams)]['goals']
```

```python
from scipy import stats

# Perform a t-test
t_stat, p_value = stats.ttest_ind(top_goals, bottom_goals)

# Print results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

P-value: The p-value is approximately 5.96e-13, which is extremely small.

# Event Analysis using hypothesis testing

Question 1: Do teams that receive more yellow cards score fewer goals?
Hypothesis:

Null Hypothesis (H0): There is no difference in the average number of goals scored by teams that receive more yellow cards.
Alternative Hypothesis (H1): Teams that receive more yellow cards score fewer goals.
Methodology:

Group the data by the number of yellow cards received and calculate the average goals scored for each group.
Perform a t-test to compare the means of the two groups (high yellow card teams vs. low yellow card teams).

```python
from scipy import stats

# Define high and low yellow card thresholds (e.g., median)
median_yellow_cards = df_merged['yellow_cards'].median()
high_yellow_cards = df_merged[df_merged['yellow_cards'] > median_yellow_cards]['goals']
low_yellow_cards = df_merged[df_merged['yellow_cards'] <= median_yellow_cards]['goals']

# Perform t-test
t_stat, p_value = stats.ttest_ind(high_yellow_cards, low_yellow_cards)

# Print results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

Interpretation:
P-value: The p-value of 0.634 is significantly higher than common significance levels (like 0.05 or 0.01). This suggests that there is insufficient evidence to reject the null hypothesis.

Conclusion:

You do not have enough statistical evidence to support the claim that teams that receive more yellow cards score fewer goals. In other words, there is no significant difference in the average number of goals scored by teams with high yellow cards compared to those with low yellow cards.

# solve using k means clustering

# Business Question
How can we segment players into different clusters based on their performance metrics (goals, assists, minutes played) and demographics (age, height, market value)?

Steps to Solve the Business Question
Data Preparation: Select relevant features and standardize them.
Apply K-means Clustering: Determine the optimal number of clusters using the Elbow method.
Visualize Clusters: Plot the clusters to understand player segmentation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Data Preparation
features = df_merged[['goals', 'assists', 'minutes_played', 'age', 'height_in_cm', 'market_value_in_eur']].dropna()

# Standardizing the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 2: Apply K-means Clustering
# Determine the optimal number of clusters using the Elbow method
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid()
plt.show()

# From the Elbow curve, choose an optimal k (e.g., k=4)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_merged['cluster'] = kmeans.fit_predict(scaled_features)


```

```python
# Step 3: Visualize Clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(x='goals', y='assists', hue='cluster', data=df_merged, palette='viridis', alpha=0.7)
plt.title('Player Clusters based on Goals and Assists')
plt.xlabel('Goals')
plt.ylabel('Assists')
plt.legend(title='Cluster')
plt.grid()
plt.show()
```

The scatter plot shows the distribution of players based on their goals and assists. The different colors represent the different clusters identified by the K-means algorithm. Cluster 0 is composed of players with low goals and assists. Cluster 1 is composed of players with high goals and assists. Cluster 2 is composed of players with low goals and high assists. Cluster 3 is composed of players with high goals and low assists

The Elbow Method plot shows the relationship between the number of clusters (k) and the inertia (within-cluster sum of squares). The optimal number of clusters is the point where the elbow occurs, meaning where the decrease in inertia starts to become less significant with additional clusters. In this case, the elbow appears to be around k=4 or k=5. This suggests that clustering the data into 4 or 5 segments might be a good choice, as adding more clusters beyond this point might not provide a significant improvement in the clustering quality.
